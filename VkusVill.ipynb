```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import time
```

```python
def timeConvert(df):
    t = pd.to_datetime(df["date_ch"], format='%Y-%m-%d')
    df["date_ch"] = t
    return df
```

```python
def sortByDays(df):
    temp = []
    for i in (pd.unique(df['id_tt_cl'])):
        t = df[df['id_tt_cl'] == i].groupby(['id_tt_cl', 'date_ch'], axis = 0).agg({'BaseSum': np.sum})#.sort_values('data_ch', ascending=False)
        temp.append(t.reset_index())
    return temp
```

```python
def weekday(df):
    temp=[]
    for i in range(len(df)):
        temp.append(df.iloc[i]['date_ch'].strftime('%w'))
    df['weekday'] = temp
    return df
```

```python
def weeksum(df):
    weekSum, num = [], []
    count = 0
    sum = 0
    for i in range(len(df)):
        if int(df.iloc[i]['weekday']) == 2:
            weekSum.append(sum)
            sum = df.iloc[i]['BaseSum']
            num.append(count)
            count += 1
        else:
            sum += df.iloc[i]['BaseSum']
    ws ={'Week numer': num, 'week sum': weekSum}
    ws = pd.DataFrame(ws)
    return ws
```
